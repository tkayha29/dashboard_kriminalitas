Let's Address the missing values

1. Let's begin by examining the variable "sex." We will explore further to identify the distinct values of sex and determine the mode or the time it occurs most frequently.

```python
unique_values = df1['Vict Sex'].unique()
print(unique_values)
```

We use M, F, and X to denote unknown or hidden values, NaN represents missing data, and H is included without a description.

To handle missing values, I will replace them with either the mode or 'X', depending on which one represents the most frequent data

Based on what we've learned, if the datatype of the fields is String or Object and there are missing values, we can fill them with the mode.

```python
modusSex = df1['Vict Sex'].mode()[0]
print(modusSex)
```

```python
df1['Vict Sex'] = df1['Vict Sex'].fillna(modusSex)
```

```python
df1.info()
```

According to data from lacity.org, victim descent encompasses a broad spectrum of racial identities in the USA. These include Black, Chinese, Korean, White, and many others.

Now let's see what the other races are

```python
unique_race = df1['Vict Descent'].unique()
print(unique_race)
```

There are 20 races, including NaN, with 'X' representing Unknown. Following the previous logic, I will use mode to substitute the missing value.

```python
modusDes = df1['Vict Descent'].mode()[0]
print(modusDes)
```

```python
df1['Vict Descent'] = df1['Vict Descent'].fillna(modusDes)
```
